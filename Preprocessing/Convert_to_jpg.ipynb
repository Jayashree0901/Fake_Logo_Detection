```python
from PIL import Image
import os

input_directory = 'G:\My Drive\New_Dataset\Fake'

output_directory = 'G:\My Drive\New_Dataset\Fake1'

os.makedirs(output_directory, exist_ok=True)

for filename in os.listdir(input_directory):
    if filename.lower().endswith(".png") or filename.lower().endswith(".jpeg"):
        with Image.open(os.path.join(input_directory, filename)) as img:
            if img.mode == 'P':
                img = img.convert('RGB')
            new_filename = os.path.splitext(filename)[0] + ".jpg"
            img.save(os.path.join(output_directory, new_filename))
            print(f"Converted {filename} to {new_filename} and saved to {output_directory}")
```
